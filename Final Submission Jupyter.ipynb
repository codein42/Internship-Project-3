```python
def barplot(x, y, data, title, color):
    plt.figure(figsize= (10, 8))
    plt.title(title)
    sns.barplot(x= x, y= y, data= data, palette= color)
    plt.show()
```

```python
task_8 = pd.DataFrame(df.groupby(['INSURANCE_TYPE'])['PREMIUM_AMOUNT'].sum())
task_8['INSURANCE_TYPE'] = task_8.index
task_8.reset_index(drop= True, inplace= True)
task_8
```

```python
df['PREMIUM_AMOUNT'].sum()
```

```python
task_8['PERCENTAGE'] = (task_8['PREMIUM_AMOUNT']/df['PREMIUM_AMOUNT'].sum())*100
```

```python
def new_premium(row):
    if ((row['INSURANCE_TYPE']=='Health') | (row['INSURANCE_TYPE']=='Property')):
        val = row['PREMIUM_AMOUNT']+(0.07*row['PREMIUM_AMOUNT'])
    elif ((row['INSURANCE_TYPE']=='Life') | (row['INSURANCE_TYPE']=='Motor')):
        val = row['PREMIUM_AMOUNT']+(0.02*row['PREMIUM_AMOUNT'])
    elif ((row['INSURANCE_TYPE']=='Mobile') | (row['INSURANCE_TYPE']=='Travel')):
        val = row['PREMIUM_AMOUNT']-(0.10*row['PREMIUM_AMOUNT'])
    return val

task_8['NEW_PREMIUM'] = task_8.apply(new_premium, axis= 1)
```

```python
task_8['NEW_PREMIUM_PRECENTAGE'] = task_8['NEW_PREMIUM']/(task_8['NEW_PREMIUM'].sum())*100
```
